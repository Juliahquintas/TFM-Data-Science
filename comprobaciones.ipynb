```python
import librosa
from pathlib import Path
import numpy as np

root = Path(r"g:\Mi unidad\TFM-Data-Science")
processed = root / "data" / "processed"

datasets = ["neurovoz", "pc-gita"]

for dataset in datasets:
    dataset_path = processed / dataset
    durations = []
    
    # Recorrer todas las carpetas
    for audio_file in dataset_path.rglob("*.wav"):
        try:
            y, sr = librosa.load(audio_file, sr=None, mono=True)
            duration = librosa.get_duration(y=y, sr=sr)
            durations.append(duration)
        except Exception as e:
            print(f"Error loading {audio_file}: {e}")
    
    if durations:
        print(f"\n{dataset}:")
        print(f"  Total audios: {len(durations)}")
        print(f"  Media duración: {np.mean(durations):.3f}s")
        print(f"  Desv. estándar: {np.std(durations):.3f}s")
        print(f"  Mín: {np.min(durations):.3f}s")
        print(f"  Máx: {np.max(durations):.3f}s")
```
