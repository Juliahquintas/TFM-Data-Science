## 2. Recorte de Silencios y Limpieza
Aquí usarás `librosa.effects.trim` para quedarte solo con la voz útil.

```python
def visualiza_audio(y, title="Forma de Onda"):
    plt.figure(figsize=(10, 3))
    librosa.display.waveshow(y, sr=16000, color='blue', alpha=0.6)
    plt.title(title)
    plt.ylabel('Amplitud')
    plt.grid(True, alpha=0.3)
    plt.show()

# visualiza_audio(y, title='Antes de recortar')
```
